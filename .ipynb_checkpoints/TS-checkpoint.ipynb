```python
import numpy as np
from scipy.stats import norm
from sympy import Symbol, diff, erf, sqrt, init_printing, ln, exp, diff, lambdify
import matplotlib.pyplot as plt
import yfinance as yf
%matplotlib inline
import random
import pandas as pd

#  Black-Scholes formulas for call and put prices
def call_price(sigma, S, K, r, t):
    small_number = 1e-6
    d1 = np.multiply(1. / sigma * np.divide(1., np.sqrt(t + small_number)),
    np.log(S/K) + (r + sigma**2 / 2.) * t)
    d2 = d1 - sigma * np.sqrt(t)
    
    C = np.multiply(S, norm.cdf(d1)) - \
        np.multiply(norm.cdf(d2) * K, np.exp(-r * t))
    return C
```

```python
# Get the ticker symbol from the user
ticker = input("Enter the ticker symbol of the stock: ")

# Get the stock data
stock = yf.Ticker(ticker)
opt = stock.option_chain(date='2023-06-16')
opt_calls = opt.calls
opt.calls

```

```python
# Select a random row from opt_calls

in_the_money = opt_calls[opt_calls['inTheMoney'] == True]
random_row = in_the_money.sample()

#####random_row = opt_calls.sample()

# Get the option data
hist = stock.history(period="1d")


# Get the expiration date from the contract symbol
expiration_str = random_row['contractSymbol'].values[0][4:10]

# Convert the expiration date to a datetime object
expiration = pd.to_datetime(expiration_str, format="%y%m%d")

S = hist["Close"].values[-1]
K = random_row['strike'].values[0]
r = 0.01
t = (expiration - pd.Timestamp.now()).days / 365
sigma = random_row['impliedVolatility'].values[0]

# Compute the call price using the Black-Scholes model
C = call_price(sigma, S, K, r, t)

# Print the option data and call price
print("Option data:")
print(random_row)
print("Underlying asset price:", S)
print("Strike price", K)
print("Expiration in years", t)
print("IV", sigma)
print("Black and Scholes - Call Option Price:", C)
```

# Symple Utility Functions
Time and spot Greeks for call options

```python
def return_time_greeks(vol, SS, KK, rr, tt):

    S = Symbol('S', real = True)
    K = Symbol('K', real = True)
    r = Symbol('r', real = True)
    t = Symbol('t', real = True)
    sigma = Symbol('sigma', real = True)

    d1 = 1 / (sigma * sqrt(t)) * (ln(S/K) + (r + sigma**2 / 2) * t)
    d2 = d1 - sigma * sqrt(t)

    C = S * 0.5 * (1 + erf(d1 / sqrt(2))) - K * exp(-r * t) * 0.5 * (1 + erf(d2 / sqrt(2)))

    Dt1 = diff(C, t)
    Dt2 = diff(Dt1, t)
    Dt3 = diff(Dt2, t)
    Dt4 = diff(Dt3, t)
    Dt5 = diff(Dt4, t)
    Dt6 = diff(Dt5, t)
    
    output0 = lambdify( (sigma, S, K, r, t), C)
    output1 = lambdify( (sigma, S, K, r, t), Dt1)
    output2 = lambdify( (sigma, S, K, r, t), Dt2)
    output3 = lambdify( (sigma, S, K, r, t), Dt3)
    output4 = lambdify( (sigma, S, K, r, t), Dt4)
    output5 = lambdify( (sigma, S, K, r, t), Dt5)
    output6 = lambdify( (sigma, S, K, r, t), Dt6)
    
    return [output0(vol, SS, KK, rr, tt), 
            output1(vol, SS, KK, rr, tt), 
            output2(vol, SS, KK, rr, tt), 
            output3(vol, SS, KK, rr, tt), 
            output4(vol, SS, KK, rr, tt), 
            output5(vol, SS, KK, rr, tt), 
            output6(vol, SS, KK, rr, tt), ]

def return_price_greeks(vol, SS, KK, rr, tt):

    S = Symbol('S', real = True)
    K = Symbol('K', real = True)
    r = Symbol('r', real = True)
    t = Symbol('t', real = True)
    sigma = Symbol('sigma', real = True)

    d1 = 1 / (sigma * sqrt(t)) * (ln(S/K) + (r + sigma**2 / 2) * t)
    d2 = d1 - sigma * sqrt(t)

    C = S * 0.5 * (1 + erf(d1 / sqrt(2))) - K * exp(-r * t) * 0.5 * (1 + erf(d2 / sqrt(2)))

    dS1 = diff(C, S)
    dS2 = diff(dS1, S)
    dS3 = diff(dS2, S)
    dS4 = diff(dS3, S)
    dS5 = diff(dS4, S)
    dS6 = diff(dS5, S)

    output0 = lambdify( (sigma, S, K, r, t), C)
    output1 = lambdify( (sigma, S, K, r, t), dS1)
    output2 = lambdify( (sigma, S, K, r, t), dS2)
    output3 = lambdify( (sigma, S, K, r, t), dS3)
    output4 = lambdify( (sigma, S, K, r, t), dS4)
    output5 = lambdify( (sigma, S, K, r, t), dS5)
    output6 = lambdify( (sigma, S, K, r, t), dS6)

    return [output0(vol, SS, KK, rr, tt), 
        output1(vol, SS, KK, rr, tt), 
        output2(vol, SS, KK, rr, tt), 
        output3(vol, SS, KK, rr, tt), 
        output4(vol, SS, KK, rr, tt), 
        output5(vol, SS, KK, rr, tt), 
        output6(vol, SS, KK, rr, tt), ]
```

# Taylor approximation using Time Greeks (Î˜, etc.)

```python
tspan = np.linspace(t, 0, 366)
C = call_price(sigma, S, K, r, tspan)

plt.plot(tspan, C, 'b', label = 'Black-Scholes')
plt.xlim(1.2*t, 0)
plt.grid(True)
plt.xlabel('Time to Expiration (Y)')
plt.ylabel('Call price ($)')

time_greeks = return_time_greeks(sigma, S, K, r, t)

C0 = time_greeks[0]
dt1 = time_greeks[1]
dt2 = time_greeks[2]
dt3 = time_greeks[3]
dt4 = time_greeks[4]
dt5 = time_greeks[5]
dt6 = time_greeks[6]

taylor = C[0] + dt1 * (tspan-t)
taylor = taylor + dt2 / np.math.factorial(2) * (tspan-t)**2
taylor = taylor + dt3 / np.math.factorial(3) * (tspan-t)**3
taylor = taylor + dt4 / np.math.factorial(4) * (tspan-t)**4
taylor = taylor + dt5 / np.math.factorial(5) * (tspan-t)**5
taylor = taylor + dt6 / np.math.factorial(6) * (tspan-t)**6

plt.plot(tspan, taylor, 'r', label = 'Taylor Approx')
plt.legend()
plt.show()

# Print the value of the Taylor approximation at the specified stock price
print(f"The Taylor approximation at t = {t/2} is {np.interp(t/2, tspan, taylor)}")
print(f"The Taylor approximation at t = {t} is {np.interp(t, tspan, taylor)}")
print(f"Black and Scholes at t = {t/2} is {np.interp(t/2, tspan, C)}")
print(f"Black and Scholes at t = {t} is {np.interp(t, tspan, taylor)}")
print(t)

# Calculate the difference between the two arrays
difference = taylor - C

# Plot the difference as a new line
plt.plot(tspan, difference, 'g', label='Difference')

plt.legend()
```

# Taylor approximation using Price Greeks (Î”, Î“, etc.)

```python
Sspan = np.linspace(S*0.8, S*1.2, 500)

C = call_price(sigma, Sspan, K, r, t)

plt.plot(Sspan, C, 'b', label = 'Black-Scholes')
plt.xlabel('Stock Price ($)')
plt.ylabel('Call price ($)')
plt.grid(True)

price_greeks = return_price_greeks(sigma, S, K, r, t)

C0 =  price_greeks[0]
dS1 = price_greeks[1]
dS2 = price_greeks[2]
dS3 = price_greeks[3]
dS4 = price_greeks[4]
dS5 = price_greeks[5]
dS6 = price_greeks[6]

taylor = C0 + dS1 * (Sspan - S)
taylor = taylor + dS2 / np.math.factorial(2) * (Sspan - S)**2
taylor = taylor + dS3 / np.math.factorial(3) * (Sspan - S)**3
taylor = taylor + dS4 / np.math.factorial(4) * (Sspan - S)**4
taylor = taylor + dS5 / np.math.factorial(5) * (Sspan - S)**5
taylor = taylor + dS6 / np.math.factorial(6) * (Sspan - S)**6

plt.plot(Sspan, taylor, 'r', label = 'Taylor Approx')
plt.legend()
plt.show()
# Find the index of the stock price closest to 200
index = np.abs(Sspan - 200).argmin()

# Print the value of the Taylor approximation at the specified stock price
print(f"The Taylor approximation at S = {S} is {np.interp(S, Sspan, taylor)}")
print(f"Black and Scholes at S = {S} is {np.interp(S, Sspan, C)}")
print(S)

# Calculate the difference between the two arrays
difference = taylor - C

# Plot the difference as a new line
plt.plot(Sspan, difference, 'g', label='Difference')

plt.legend()
```
